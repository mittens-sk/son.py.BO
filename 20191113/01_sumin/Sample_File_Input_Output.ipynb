```python
def outputDatFile(output, seq, yymmdd = None, colTF = False): 
    
    inputPrefix = "SON_TO_TEOS_RL_REQUEST_"
    
    if yymmdd is None:
        yymmdd = datetime.today().strftime("%Y%m%d")

    nm = inputPrefix + str(yymmdd) + "_" + str(seq) + ".dat"

    try :
#          output.to_csv(nm, sep = "|", header = False, index = False)        
         output.to_csv(nm, sep = "|", header = colTF, index = False)        


    except FileNotFoundError:
        pass
    
```

```python
outputDatFile(output = df, seq = 11, yymmdd = '20191002', colTF = True) # including column name
outputDatFile(output = df, seq = 21, yymmdd = '20191002', colTF = False) 
```
