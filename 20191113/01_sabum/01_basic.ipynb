```python
# Scikit Learn Data 불러와서 pandas dataframe로 변환
b=sk_data.load_boston()
X=pd.DataFrame(b.data, columns=b.feature_names)
Y=pd.DataFrame(b.target, columns=['MEDV'])
df=pd.concat([X, Y], axis=1)
columns=df.columns
index=df.index
data=df.values
type(index)
```
